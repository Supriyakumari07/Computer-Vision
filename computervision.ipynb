```python
def filtering(img,f=3):
    
    (rows,col,channels)=img.shape
    stride=2
    
    n_row=int(1+(rows-f)/stride)
    n_col=int(1+(col-f)/stride)
    n_channels=channels

    n_img=np.zeros((n_row,n_col,n_channels))
    
    for h in range(n_row):
        for w in range(n_col):
            for c in range(n_channels):
                vert_start=h*stride
                vert_end=vert_start+f
                horiz_start=w*stride
                horiz_end=horiz_start+f
                n_slice=img[vert_start : vert_end, horiz_start: horiz_end,c]
                n_img[h,w,c]=np.mean(n_slice, dtype=int)
    return n_img
```

```python
A=filtering(img)
print(A.shape)
plt.imshow(img)
plt.show()
```

```python
A=filtering(img,f=25)
print(A.shape)
plt.imshow(img)
plt.show()
```
